#### IMPORTACIÓN DE LIBRERIAS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### LECTURA Y LIMPIEZA DE DATOS

```python
df=pd.read_csv("listings.csv", encoding="UTF-8-sig")
```

```python
df.info()
```

```python
df.dropna(axis=1, how="all", inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.loc[df.last_review.isnull()]
```

```python
df.loc[df.reviews_per_month.isnull()]
```

```python
df.loc[df.price.isnull(), "price"]=df.price.median()
```

```python
df1 = df.copy()
```

```python
from scipy.stats.mstats import winsorize
df["price"] = winsorize(df["price"], limits=[0.05,0.05])
df["minimum_nights"] = winsorize(df["minimum_nights"], limits=[0.05,0.05])
#df.head(20)
```

```python
df["last_review"] = pd.to_datetime(df["last_review"])
```

#### COMPARACIÓN DEL ANTES Y DESPUES DEL MANEJO DE OUTLIERS

```python
df.boxplot(["price","minimum_nights"])
plt.xlabel("Columnas")
plt.ylabel("Valores")
plt.title("Caja de Bigotes de las Columnas Precios y Noches Mínimas Después del Manejo de Outliers")
plt.show()
```

```python
df1.boxplot(["price","minimum_nights"])
plt.xlabel("Columnas")
plt.ylabel("Valores")
plt.title("Caja de Bigotes de las Columnas Precios y Noches Mínimas Antes del Manejo de Outliers")
plt.show()
```
